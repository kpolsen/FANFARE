# Tutorial 4 -  iii) Storage capacity requirements from DFT analysis
## Step 1: Import FANFARE source code

```python
import source as sc
```

## Step 2: Load data
Here we load a dataset downloaded from Energidataservice.dk (see Tutorial 0):

```python
data_ob = sc.PowerData(load=True,file_name='energinet_all',year='2018')
```

## Step 3: Decide on frequency setup
Determine the frequency intervals in terms of timescales in hours, and write an equal number of labels to go with them. **Note that the longest timescale is cut of at 1 yr, due to the chosen time span of the data**:

```python
duration_cuts = [[0.25,5],[5,24],[24,7*24],[7*24,3*30*24],[3*30*24,365*24]] # hours
labels = ['< 5 hrs',\
          '5 - 24 hrs',\
          '24 hrs - 1 week',\
          '1 week - 3 months',\
          '3 months - 1 yr']
labels_verbose = ['Hourly','Intra-daily','Daily','Weekly','Seasonally']
data_ob.SetFrequencySetup(duration_cuts=duration_cuts,labels=labels)
```

## Step 4: Derive storage capacity requirements and plot as box plot

```python
capacity_DK = data_ob.GetCapacityReq(col_name='ResLoad_DK',plot=True,verbose=True)
```
