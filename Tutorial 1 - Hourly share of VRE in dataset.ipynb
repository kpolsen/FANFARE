# Tutorial 1 - Hourly share of VRE in dataset
## Step 1: Import FANFARE source code

```python
import source as sc
```

## Step 2: Load data
Here we load a dataset downloaded from Energidataservice.dk (see Tutorial 0):

```python
data_ob = sc.PowerData(load=True,file_name='energinet_all',year='2018')
```

## Step 3: Add hourly VRE share to data

```python
alpha_cuts = [20,80]
data_ob.AddHighPenetrationIndex(alpha_cuts=alpha_cuts,power_name='RenPower_DK',load_name='GrossCon_DK')
data_ob.AddHighPenetrationIndex(alpha_cuts=alpha_cuts,power_name='RenPower_DK1',load_name='GrossCon_DK1')
data_ob.AddHighPenetrationIndex(alpha_cuts=alpha_cuts,power_name='RenPower_DK2',load_name='GrossCon_DK2')
```

```python
data_ob.PlotHisto(bins=100,legend=True,\
           histogram_values=['alpha_'+'RenPower_DK','alpha_'+'RenPower_DK1','alpha_'+'RenPower_DK2'],\
           colors=['k','slategrey','blue'],ls=['-','--',':'],ylab='Time spent [%]',\
           labels=['DK 2018 data','DK1 2018 data','DK2 2018 data'],\
           xlim=[0,150],xlab=r'Hourly share of VRE, $\alpha_{\mathrm{VRE}}$ [%]')
```
